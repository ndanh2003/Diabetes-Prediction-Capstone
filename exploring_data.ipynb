```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("diabetes_prediction_dataset.csv")
data = pd.get_dummies(data, columns=["gender", "smoking_history"], drop_first=True, dtype=float)
for c in data.columns.drop('diabetes'):
    data.plot.scatter(x=c,y='diabetes',alpha=0.05)

plt.show()
```
